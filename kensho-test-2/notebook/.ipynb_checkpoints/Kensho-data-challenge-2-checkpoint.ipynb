## Feature Engineering

This part is done iteratively

Basic Engineering

```python
# drop rows with (null) Borough
drop_row = (X_train["Borough"] != "(null)")
X_train = X_train[drop_row]
y_train = y_train[drop_row]
```

Since it is a short-time challenge, I will drop the data which requires more time to work with

```python
X_train.columns
```

```python
drop_col_more = ['XCoordinate', 'YCoordinate', 'Occurrence Date', 'CompStat Year', 'Occurrence Year']
```

```python
X_train.drop(drop_col_more, axis=1, inplace=True)
X_test.drop(drop_col_more, axis=1, inplace=True)
```

```python
X_train.columns
```

We are working with categorical features, so we have to encode them before modeling

Basic encoding

```python
lb = LabelBinarizer()
```

```python
DoW_train = pd.DataFrame(lb.fit_transform(X_train["Day of Week"]),
             columns=["Day_of_Week_{}".format(i) for i in range(X_train["Day of Week"].value_counts().shape[0])])

DoW_test = pd.DataFrame(lb.fit_transform(X_test["Day of Week"]),
             columns=["Day_of_Week_{}".format(i) for i in range(X_test["Day of Week"].value_counts().shape[0])])
```

```python
Hr_train = pd.DataFrame(lb.fit_transform(X_train["Occurrence Hour"]),
             columns=["Occurrence_Hour_{}".format(i) for i in range(X_train["Occurrence Hour"].value_counts().shape[0])])

Hr_test = pd.DataFrame(lb.fit_transform(X_test["Occurrence Hour"]),
             columns=["Occurrence_Hour_{}".format(i) for i in range(X_test["Occurrence Hour"].value_counts().shape[0])])
```

```python
Br_train = pd.DataFrame(lb.fit_transform(X_train["Borough"]),
             columns=["Borough_{}".format(i) for i in range(X_train["Borough"].value_counts().shape[0])])

Br_test = pd.DataFrame(lb.fit_transform(X_test["Borough"]),
             columns=["Borough_{}".format(i) for i in range(X_test["Borough"].value_counts().shape[0])])
```

Encoding and engineering

I will transform the data before encoding

For CompStat Day, I will cut the day of month into 3 ranges (1,10), (11,20) and (20,30)

```python
D_train = pd.cut(X_train["CompStat Day"], bins=[0, 10, 20, 31], right=True, labels=[0,1,2])
D_train = pd.DataFrame(lb.fit_transform(D_train),
             columns=["Day_{}".format(i) for i in range(D_train.value_counts().shape[0])])

D_test = pd.cut(X_test["CompStat Day"], bins=[0, 10, 20, 31], right=True, labels=[0,1,2])
D_test = pd.DataFrame(lb.fit_transform(D_test),
             columns=["Day_{}".format(i) for i in range(D_test.value_counts().shape[0])])
```

```python
assert D_train.shape[1] == D_test.shape[1]
```

For Sector, I will use the top 10 major categories and group the left to "others" category

```python
sector_major = list(X_train["Sector"].value_counts().sort_values(ascending=False).index[:10])
```

```python
Sc_train = X_train["Sector"].apply(lambda x: x if x in sector_major else "others")

Sc_train = pd.DataFrame(lb.fit_transform(Sc_train),
             columns=["Sc_{}".format(i) for i in range(Sc_train.value_counts().shape[0])])
```

```python
Sc_test = X_test["Sector"].apply(lambda x: x if x in sector_major else "others")

Sc_test = pd.DataFrame(lb.fit_transform(Sc_test),
             columns=["Sc_{}".format(i) for i in range(Sc_test.value_counts().shape[0])])
```

```python
assert Sc_train.shape[1] == Sc_test.shape[1]
```

For Precinct, to be safe, I will only create an indicator of the top major category

```python
Pr_major = X_train["Precinct"].value_counts().sort_values(ascending=False).index[0]

Pr_train = (X_train["Precinct"] == Pr_major) * 1
Pr_test = (X_test["Precinct"] == Pr_major) * 1
```

For Jurisdiction, I will extract top 3 major classes

```python
Ju_major = X_train["Jurisdiction"].value_counts().sort_values(ascending=False).index[:3]
```

```python
Ju_train = X_train["Jurisdiction"].apply(lambda x: x if x in Ju_major else "others")

Ju_train = pd.DataFrame(lb.fit_transform(Ju_train),
             columns=["Ju_{}".format(i) for i in range(Ju_train.value_counts().shape[0])])
```

```python
Ju_test = X_test["Jurisdiction"].apply(lambda x: x if x in Ju_major else "others")

Ju_test = pd.DataFrame(lb.fit_transform(Ju_test),
             columns=["Ju_{}".format(i) for i in range(Ju_test.value_counts().shape[0])])
```

```python
assert Ju_train.shape[1] == Ju_test.shape[1]
```

#### Combine the engineered data

```python
X_train_eg = np.concatenate([DoW_train.as_matrix(), Hr_train.as_matrix(), Br_train.as_matrix()
                             , D_train.as_matrix(), Sc_train.as_matrix(), Pr_train.as_matrix().reshape(-1,1), 
                             Ju_train.as_matrix()], axis=1)
```

```python
X_train_eg.shape
```

```python
X_test_eg = np.concatenate([DoW_test.as_matrix(), Hr_test.as_matrix(), Br_test.as_matrix()
                             , D_test.as_matrix(), Sc_test.as_matrix(), Pr_test.as_matrix().reshape(-1,1), 
                             Ju_test.as_matrix()], axis=1)
```

```python
X_test_eg.shape
```

We have finished the basic engineering and could start to split the data and build the model

## Train and Validation set split --- Stratefied

Now split the data into 2 data set <br>
1. train
2. validation

Only use the data of train and validation for model training<br>
Use the test to evaluate the generalization of the final model

```python
X_tr, X_val, y_tr, y_val = train_test_split(X_train_eg, y_train, test_size=0.2, random_state=18, stratify=y_train)
```

Transform y from str to int

```python
ecd = LabelEncoder()
ecd.fit(y_tr)
```

```python
y_tr = ecd.transform(y_tr)
y_val = ecd.transform(y_val)
y_test = ecd.transform(y_test)
```

## Model prototyping

Since our data set contains multiple binary encoding data, it is better to start from the ***tree based model***

I will only experiment with gradient boosting tree based on the time limitation

```python
params = {'eta': 0.08, 'max_depth': 4, 'subsample': 0.8, 'colsample_bytree': 0.8, "num_class": 7,
          'objective': 'multi:softmax', 'silent': True}

num_round = 500
```

```python
xg_train = xgb.DMatrix(X_tr, label=y_tr)
xg_val = xgb.DMatrix(X_val, label=y_val)
```

```python
watchlist = [(xg_train, 'train'), (xg_val, 'val')]
```

```python
def acc(preds, dtrain):
    labels = dtrain.get_label()
    acc_score = accuracy_score(labels, preds, normalize=True, sample_weight=None)
    return 'acc', acc_score
```

```python
bst = xgb.train(params, xg_train, num_round, 
                watchlist, early_stopping_rounds=50, 
                feval=acc, maximize=True, verbose_eval=20)
```

### Error Analysis

The model seems perform not very well, only surpass the dummy model for about 3%, what happened?

```python
# get prediction
val_pred = bst.predict(xg_val)
accuracy_score(y_val, val_pred, normalize=True, sample_weight=None)
```

```python
# Compute confusion matrix
cnf_matrix = confusion_matrix(y_val, val_pred)
np.set_printoptions(precision=2)

# Plot confusion matrix
plt.figure(figsize=(4,4));
plot_confusion_matrix(cnf_matrix, classes = [0,1,2,3,4,5,6], title='Confusion matrix')
print(ecd.classes_)
```

1. We can see that we missed a lot on the prediction of class 3,4,5, all of them are mis-classified to be the major crime: GRAND LARCENY. 
2. The classification of 'BURGLARY', 'FELONY ASSAULT' seems to be hard to classify because they share similar mis-classification problems. We should dive into data more to dig out the features for class 'GRAND LARCENY OF MOTOR VEHICLE', 'MURDER & NON-NEGL. MANSLAUGHTE','RAPE'. They are minor classes and will be seriously influenced by the major class
3. Class 6 ROBBERY seems to be easier to classify, but still missed a lot on the 3 major classes
