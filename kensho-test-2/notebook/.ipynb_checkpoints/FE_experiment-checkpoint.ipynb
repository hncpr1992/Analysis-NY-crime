```python
def recon(reg):
    integer = int(np.round((40*reg)**2)) 
    for a in range(32):
        if (integer - a) % 31 == 0:
            A = a
    M = (integer - A)//31
    return A, M


X['ps_reg_A'] = X['ps_reg_03'].apply(lambda x: recon(x)[0])
X['ps_reg_M'] = X['ps_reg_03'].apply(lambda x: recon(x)[1])
X['ps_reg_A'].replace(19,-1, inplace=True)
X['ps_reg_M'].replace(51,-1, inplace=True)
```

```python
def create_interactions(data, feat_list):
    for comb in feat_list:
        feat = comb[0] + "_" + comb[1]
        new_feat = data[comb[0]].astype(str) + "+" + data[comb[1]].astype(str)
        new_feat = pd.DataFrame(lb.fit_transform(new_feat), 
                                columns=[(feat+"_{}").format(i+1) 
                                for i in range(new_feat.value_counts().shape[0])])
        data = pd.concat([data, new_feat], axis=1)
    return data

feat_list = [
             ['ps_reg_01', 'ps_car_02_cat'],  
             ['ps_reg_01', 'ps_car_04_cat']
]

X = create_interactions(X.copy(), feat_list)
```

## model running

```python
dtest = xgb.DMatrix(data_test.as_matrix())
```

```python
def gini(actual, pred, cmpcol = 0, sortcol = 1):
    assert( len(actual) == len(pred) )
    all = np.asarray(np.c_[ actual, pred, np.arange(len(actual)) ], dtype=np.float)
    all = all[ np.lexsort((all[:,2], -1*all[:,1])) ]
    totalLosses = all[:,0].sum()
    giniSum = all[:,0].cumsum().sum() / totalLosses
    giniSum -= (len(actual) + 1) / 2.
    return giniSum / len(actual)
 
def gini_normalized(a, p):
    return gini(a, p) / gini(a, a)

def gini_xgb(preds, dtrain):
    labels = dtrain.get_label()
    gini_score = gini_normalized(labels, preds)
    return 'gini', gini_score
```

```python
X_in = X.as_matrix()
y_in = y.as_matrix()
```

```python
# params 1
# params = {'eta': 0.015, 'max_depth': 5, 'subsample': 0.9, 'colsample_bytree': 0.9, 
#           'objective': 'binary:logistic', 'eval_metric': 'auc', 'silent': True}

# params 2 (best)
params = {'eta': 0.07, 'max_depth': 4, 'subsample': 0.8, 'colsample_bytree': 0.8, 
          "min_child_weight":6,"gamma":10,"reg_alpha":8,"reg_lambda":1.3,"scale_pos_weight":1.6,
          'objective': 'binary:logistic', 'eval_metric': 'auc', 'silent': True}
```

```python
# np.random.seed(123)
kfolds = 5
pred = np.zeros([data_test.shape[0], 5])

# del X
# del data_test

# stratefied
skf = StratifiedKFold(n_splits=kfolds)
# model_collection = []

for i, (train_index, valid_index) in enumerate(skf.split(X_in, y_in)):
    print("The {} fold".format(i))
    print("###################################################")
    
    # train test split
    X_train, X_valid = X_in[train_index], X_in[valid_index]
    y_train, y_valid = y_in[train_index], y_in[valid_index]
    dtrain = xgb.DMatrix(X_train, label=y_train.reshape(-1,1))
    dvalid = xgb.DMatrix(X_valid, label=y_valid.reshape(-1,1))
    evallist = [(dtrain, 'train'), (dvalid, 'valid')]
    
    mdl = xgb.train(params, dtrain, 
                600, evallist, early_stopping_rounds=100, 
                feval=gini_xgb, maximize=True, verbose_eval=50)
    pred[:, i] = mdl.predict(dtest, ntree_limit=mdl.best_ntree_limit)
```
