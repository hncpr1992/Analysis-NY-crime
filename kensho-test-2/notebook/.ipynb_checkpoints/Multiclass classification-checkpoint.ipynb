## Model construction

Here, we will consider to use the xgboost, which is better than random forest and neural network

We will not consider stacking multiple model in this short-time project, only ensemble a single model

The cross-entropy from xgboost is 0.209302, we need to improve it with fine tuning and ensemble

Ensemble

```python
dtest = xgb.DMatrix(X_test)
```

```python
params = {'eta': 0.1, 'max_depth': 4, 'subsample': 0.8, 'colsample_bytree': 0.8, "num_class": 6,
          'objective': 'multi:softprob', 'silent': True}
```

```python
# np.random.seed(123)
kfolds = 2
pred = np.zeros([y_test.shape[0], 6])
```

```python
def f1_sc(preds, dtrain):
    labels = dtrain.get_label()
    f1 = f1_score(labels, preds.argmax(axis=1), average="macro")
    return 'f1-macro', f1
```

```python
# stratefied
skf = StratifiedKFold(n_splits=kfolds)
# model_collection = []

for i, (train_index, valid_index) in enumerate(skf.split(X_train, y_train.as_matrix())):
    print("The {} fold".format(i+1))
    print("###################################################")
    
    # train test split
    X_tr, X_va = X_train[train_index], X_train[valid_index]
    y_tr, y_va = y_train.as_matrix()[train_index], y_train.as_matrix()[valid_index]
    dtr = xgb.DMatrix(X_tr, label=y_tr)
    dva = xgb.DMatrix(X_va, label=y_va)
    watchlist = [(dtr, 'train'), (dva, 'valid')]

    bst = xgb.train(params, dtr, 500, 
                watchlist, feval=f1_sc, maximize=True, early_stopping_rounds=20, verbose_eval=10)
    pred += bst.predict(xg_val, ntree_limit=bst.best_ntree_limit) / kfolds
```

```python
f1_score(y_val, pred.argmax(axis=1), average="macro")
```

```python
# np.random.seed(123)
kfolds = 2
pred = np.zeros([y_test.shape[0], 6])
```

```python
# stratefied
skf = StratifiedKFold(n_splits=kfolds)
# model_collection = []

for i, (train_index, valid_index) in enumerate(skf.split(X_model, y_model.as_matrix())):
    print("The {} fold".format(i+1))
    print("###################################################")
    
    # train test split
    X_tr, X_va = X_model[train_index], X_model[valid_index]
    y_tr, y_va = y_model.as_matrix()[train_index], y_model.as_matrix()[valid_index]
    dtr = xgb.DMatrix(X_tr, label=y_tr)
    dva = xgb.DMatrix(X_va, label=y_va)
    watchlist = [(dtr, 'train'), (dva, 'valid')]

    bst = xgb.train(params, dtr, 500, 
                watchlist, feval=f1_sc, maximize=True, early_stopping_rounds=20, verbose_eval=10)
    pred += bst.predict(dtest, ntree_limit=bst.best_ntree_limit) / kfolds
```

```python
f1_score(y_test, pred.argmax(axis=1), average="macro")
```
